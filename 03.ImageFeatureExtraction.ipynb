# Extracting Features from Images
##### Obtaining pixel intensities for image and transforming those into usable features

- Images are represented as a matrix
- Every cell in the matrix represents a particular pixel in the image
- Each pixel holds a value based on the type of image
- For a color image every is pixel is represented using 3 values corresponding to R,G,B: 0-255 (ex: red(255,0,0), green(0,255,0), blue(0,0,255))
- Color images are called 3 channel images
- Grayscale Images require just one value for each pixel, this value represents the intensity of the pixel, a number between 0 - 1
- Grayscale images are called 1 channel images
- Both color & Grayscale Images can be represented by a 3D matrix the first two dimensions represent the height & width of the image the third dimension represents the number of channels of the image
- scikit-image is a collection of algorithms for image processing
