# Linear, Lasso and Ridge Regression

Regression
- is one of the most common use case for machine learning
- is where we try to predict an output value (Dependent variable) from a continuous range based on input features (Explanatory variables, Independent variables)
- Residuals of a regression are the difference between actual values and fitted values of the dependent variable
- The best fit regression line is the one where the sum of the squares of the lengths of the residuals/error is minimum, this approach is called Ordinary Least Square Regression (OLS)
- MSE mean Square error is the variance of the residuals 
- When we solve the linear regression problem using the OLS method we get the "Best Linear Unbiased Estimator (BLUE)"
  - it is called the Best because the COefficients have minimum variance, i.e. are estimated with relatively high certainty
  - it is called Unbiased because the residuals have zero mean, are uncorrelated to each other and have equal variance
- To measure how good a regression is we use metric called R<sup>2</sup> = ESS / TSS (ESS: Explained Sum of Squares(Variance of fitted values), TSS: Total Sum of Squares(Variance of actual values))
- Usualy higher the value of R2 the better is the quality of the regression (upper bound 1)
- R<sup>2</sup> tries to measure how much of the original variance is captured in the fitted values
- Other Types of Regression are
  - Ridge Regression
  - Lasso Regression
  - Elastic Net Regression
  - SVR Regression

#### Since there are only 7 unique values, we can explicitly set the corresponding numeric values

```python
cylinders_dict = {'two': 2, 
                  'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}
auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)

auto_data.head()
```

#### All other non-numeric fields can be made into usable features by applying one-hot-encoding

```python
auto_data = pd.get_dummies(auto_data, 
                           columns=['make', 
                                    'fuel-type', 
                                    'aspiration', 
                                    'num-of-doors', 
                                    'body-style', 
                                    'drive-wheels', 
                                    'engine-location', 
                                    'engine-type', 
                                    'fuel-system'])
auto_data.head()
```

#### Drop rows containing missing values

```python
auto_data = auto_data.dropna()
auto_data
```

#### Verify that there are no null values in the data set

```python
auto_data[auto_data.isnull().any(axis=1)]
```

### Data Cleaning is now complete
We can now use our data to build our models

#### Create training and test data using train_test_split

```python
from sklearn.model_selection import train_test_split

# Independent variables
X = auto_data.drop('price', axis=1)

# Taking the labels (price)
Y = auto_data['price']

# Spliting into 80% for training set and 20% for testing set so we can see our accuracy
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

#### Check R<sup>2</sup> on training data

```python
linear_model.score(X_train, Y_train)
```

#### View coefficients for each feature

```python
linear_model.coef_
```

#### A better view of the coefficients
List of features and their coefficients, ordered by coefficient value

```python
predictors = X_train.columns
coef = pd.Series(linear_model.coef_,predictors).sort_values()

print(coef)
```

#### Make predictions on test data

```python
y_predict = linear_model.predict(x_test)
```

#### Compare predicted and actual values of Price

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 5)

plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('Price')

plt.legend()
plt.show()
```

#### R<sup>2</sup> score
For our model, how well do the features describe the price?

```python
r_square = linear_model.score(x_test, y_test)
r_square
```

#### Calculate Mean Square Error

```python
from sklearn.metrics import mean_squared_error

linear_model_mse = mean_squared_error(y_predict, y_test)
linear_model_mse
```

#### Root of Mean Square Error

```python
import math

math.sqrt(linear_model_mse)
```

the RMSE value in this case indicates that the predicted price of the car may differ by about 5108 $ on an average from the actual price either in +ve or -ve direction.

There are different ways in which we can measure distance between two points like:
- L1 distance:
  - Aslo called (Snake distance, City block distance, Manhattan distance) 
  - Is calculated by finding the absolute difference between the cordinates of the two points & then summing these differences: sum(abs(t1-t2))
  - L1 Norm for any set of numbers is the absolute value of those numbers, the term 1 in the name L1 comes from the fact that all numbers are raised to the power of 1
  - L2 Norm is the absolute value of those numbers with each one of them raised to the power of 2

- L2 distance: (Euclidean Distance)
  - sqrt(sum(t1-t2)2)

Bias-Variance Trade-off
- Low bias means fewer assumptions about the data & fewer parameter constraints on the model
- The model which has low bias has more complex model the model parameters count for little & hence more dependent on training data thus tend to overfit
- High bias means more assumptions about the data 
- The model which has high bias has too simple model the model parameters become far more important than the training data thus tend to underfit
- Low Variance model is one which does not change much when the training data changes, it is not very sensitive to the data points in the training data
- Low Variance model can be too simple because it is not much sensitive to the training data
- High Variance model is one which changes significantly as the training data changes, it is highly sensitive to the training data
- High Variance model might be too complex because it depends a lot on the training data
- A model which is too complex is said to have High Variance error
- A model which is too simple is said to High bias error
- An overfitted model has:
  - High variance error
  - Low bias error
- While chosing the model depending on the use case & the training data at hand we need to have the correct Bias-Variance Trade-off
- High-bias algoriths have simple parameters example are Regression
- High-Variance algorithms have complex parameters example are Decision Trees, Dense Neural Networks


Overfitting is common problem with machine learning alrgorithms thus there are different ways to Prevent Overfitting namely:
- Regularization
- Cross-validation
- Ensemble learning
  - Dropout

- Regularization: 
  - is a technique where we Penalize complex models, is used by Lasso, Ridge, Elastic Net Regression
  - Reduces Variance error but Increases Bias error
- Cross-validation: we split the training data into distinct training & validation sets
- Dropouts: are in Neural networks only in which we intentionally trn off some neurons during training (i.e. different neuron disabled for every epoch of training)

- A big risk with multiple variable regression is multicollinearity i.e. the X variables which contain the same information, a feature which is very highly correlated can cause overfitting

Lasso Regression (Lasso -> Least Absolute Shrinkage and Selection Operator)
- Penalizes large regression coefficients
- Uses a penalty function with L1 norms of the coefficients added to the MSE, penalty is weighted by hyperparameter alpha(alpha helps eliminate unimportant features)
- Very High alpha value will force some coefficients to converge to 0

Ridge Regression
- Penalizes large regression coefficients
- Uses a penalty function with L2 norms of the coefficients added to the MSE, penalty is weighted by hyperparameter alpha
- Unlike Lasso, Ridge regression has closed-form solution i.e. it can. be expressed in terms of a formula
- Unlike Lasso, Ridge regression will not force coefficients to 0 if tune the alpha parameter to very high value

Elastic Net Regression
- Simply combines Lasso & ridge

### Lasso Regression
Cost Function: RSS + <b>&alpha;</b>*(sum of absolute values of coefficients)

RSS = Residual Sum of Squares

Larger values of <b>&alpha;</b> should result in smaller coefficients as the cost function needs to be minimized

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=0.5, normalize=True)
lasso_model.fit(X_train, Y_train)
```

```python
lasso_model.score(X_train, Y_train)
```

```python
y_predict = lasso_model.predict(x_test)
```

RSS = Residual Sum of Squares
